```python
import pandas as pd # Import the pandas library for data manipulation
import numpy as np # Import the numpy library for numerical operations
import matplotlib.pyplot as plt # Import the matplotlib library for data visualization
import seaborn as sns # Import the seaborn library for advanced data visualization

# Load dataset
titanic = pd.read_csv("titanic.csv")
# titanic.head(5)

# Data preprocessing
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].mean()) # Fill missing values with median age 
titanic['Cabin'] = titanic['Cabin'].fillna('Unknown')
titanic['Embarked'] = titanic['Embarked'].fillna(data["Embarked"].mode())

# Set style
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))

# 1. Basic Distributions
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Age distribution
sns.histplot(titanic['Age'], bins=50, kde=True, ax=axes[0,0], color='red',multiple="stack")
axes[0,0].set_title('Age Distribution')

# Gender distribution
sns.countplot(x='Sex', data=titanic, ax=axes[0,1], palette='pastel',hue='Survived')
axes[0,1].set_title('Gender Distribution')

# Survival count
sns.countplot(x='Survived', data=titanic, ax=axes[1,0], palette='Set2')
axes[1,0].set_title('Survival Count (0 = Died, 1 = Survived)')

# Passenger class distribution
sns.countplot(x='Pclass', data=titanic, ax=axes[1,1], palette='Paired')
axes[1,1].set_title('Passenger Class Distribution')

plt.tight_layout()
plt.show()

# 2. Survival Analysis
plt.figure(figsize=(15, 10))

# Survival by gender
plt.subplot(2, 2, 1) # 2 rows, 2 columns, first subplot
sns.countplot(x='Survived', hue='Sex', data=titanic, palette='coolwarm')
plt.title('Survival by Gender')

# Survival rate by passenger class
plt.subplot(2, 2, 2) # 2 rows, 2 columns, second subplot
sns.barplot(x='Pclass', y='Survived', data=titanic, palette='viridis')
plt.title('Survival Rate by Passenger Class')

# Age distribution of survivors vs non-survivors
plt.subplot(2, 2, 3) # 2 rows, 2 columns, third subplot
sns.kdeplot(titanic[titanic['Survived'] == 1]['Age'], label='Survived', shade=True)
sns.kdeplot(titanic[titanic['Survived'] == 0]['Age'], label='Died', shade=True)
plt.title('Age Distribution: Survivors vs Non-Survivors')

# Fare vs Survival
plt.subplot(2, 2, 4) # 2 rows, 2 columns, fourth subplot
sns.boxplot(x='Survived', y='Fare', data=titanic, palette='Set3')
plt.title('Fare Distribution: Survivors vs Non-Survivors')

plt.tight_layout()
plt.show()

# 3. Advanced Analysis
plt.figure(figsize=(18, 12))

# Correlation heatmap
plt.subplot(2, 2, 1)
corr = titanic[['Survived','Pclass','Age','SibSp','Parch','Fare']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')

# Family size impact
titanic['FamilySize'] = titanic['SibSp'] + titanic['Parch']
plt.subplot(2, 2, 2)
sns.pointplot(x='FamilySize', y='Survived', data=titanic, color='purple')
plt.title('Survival Rate by Family Size')

# Age distribution across classes
plt.subplot(2, 2, 3)
sns.violinplot(x='Pclass', y='Age', hue='Survived', data=titanic, split=True, palette='magma')
plt.title('Age Distribution Across Classes by Survival')

# Embarkation analysis
plt.subplot(2, 2, 4)
sns.barplot(x='Embarked', y='Survived', hue='Pclass', data=titanic, palette='Set2')
plt.title('Survival Rate by Embarkation Port and Class')

plt.tight_layout()
plt.show()

# 4. Categorical Relationships
plt.figure(figsize=(15, 10))

# Survival rate by class and gender
plt.subplot(2, 2, 1)
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=titanic, palette='rocket')
plt.title('Survival Rate by Class and Gender')

# Fare distribution by class
plt.subplot(2, 2, 2)
sns.boxenplot(x='Pclass', y='Fare', data=titanic, palette='dark')
plt.title('Fare Distribution by Class')

# Age distribution by survival and class
plt.subplot(2, 2, 3)
sns.swarmplot(x='Pclass', y='Age', hue='Survived', data=titanic, palette='Set1')
plt.title('Age Distribution by Class and Survival')

plt.tight_layout()
plt.show()
```

```python
from scipy import stats
def chi2_test(feature):
    contingency_table = pd.crosstab(data[feature], data['Survived'])
    chi2, p, _, _ = stats.chi2_contingency(contingency_table)
    print(f"{feature}: χ² = {chi2:.2f}, p = {p:.4f}")

chi2_test('Gen')
chi2_test('Pclass')
chi2_test('Embarked')
```

```python
survived = data[data['Survived'] == 1]
not_survived = data[data['Survived'] == 0]

for feature in ['Age', 'Fare']:
    t_stat, p_val = stats.ttest_ind(survived[feature], 
                                  not_survived[feature],
                                  equal_var=False)
    print(f"{feature}: t = {t_stat:.2f}, p = {p_val:.4f}")
```
